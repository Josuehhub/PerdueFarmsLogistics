```python
#Discrption: 5 Questions for perdue farm 
#
#
#

import gspread
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def tms():
    df_tms = pd.read_csv (r'/Users/josueherrera/Desktop/Towson University/Fall 2021/Data Anayltics Competition/Perdue Farms/data/csv_folder/TMS.csv')

    #change type
    df_tms [['Pickup Depart Date']] = df_tms[['Pickup Depart Date']].apply(pd.to_datetime)
    df_tms['Pickup Depart Time'] = pd.to_timedelta(df_tms['Pickup Depart Time'])

    #Create Datetime column
    df_tms['Pickup Depart DateTime'] = pd.to_datetime(df_tms['Pickup Depart Date'] + df_tms['Pickup Depart Time'])

    #Reorder
    df_tms = df_tms.drop(df_tms.iloc[:, 6:8], axis=1)

    return(df_tms)
tms()
```

```python
def time_csv():

    df_time = pd.read_csv (r'/Users/josueherrera/Desktop/Towson University/Fall 2021/Data Anayltics Competition/Perdue Farms/data/csv_folder/Time.csv')

    #Change data type
    df_time[['Empty Date', 'Actual Arrive Date', 'Sched Arrive Date']] = df_time[['Empty Date', 
                                                                                  'Actual Arrive Date', 
                                                                                  'Sched Arrive Date']].apply(pd.to_datetime)


    df_time['Sched Arrive Time'] = pd.to_timedelta(df_time['Sched Arrive Time'])
    df_time['Empty Time'] = pd.to_timedelta(df_time['Empty Time'])
    df_time['Actual Arrive Time'] = pd.to_timedelta(df_time['Actual Arrive Time'])

    df_time['LATE? (Yes/No)'] = df_time['LATE? (Yes/No)'].astype('object')
    df_time['Sold To'] = df_time['Sold To'].astype('int')

    #Create Datetime column
    df_time['Sched Arrive DateTime'] = pd.to_datetime(df_time['Sched Arrive Date'] + df_time['Sched Arrive Time'])
    df_time['Actual Arrive DateTime'] = pd.to_datetime(df_time['Actual Arrive Date'] + df_time['Actual Arrive Time'])
    df_time['Empty DateTime'] = pd.to_datetime(df_time['Empty Date'] + df_time['Empty Time'])

    #Reorder Columns/ remove Columns
    df_time = df_time[['Sold To', 'Carrier Name', 'Load #', 'Sched Arrive DateTime', 'Actual Arrive DateTime', 
        'Empty DateTime', 'LATE? (Yes/No)', 'HELD']]

    df_time['HELD'] = df_time['Empty DateTime'] - df_time['Actual Arrive DateTime']

    df_time.loc[df_time['Actual Arrive DateTime'] - df_time['Sched Arrive DateTime'] > '00:30:00', 'LATE? (Yes/No)'] = "Yes"
    df_time.loc[df_time['Actual Arrive DateTime'] - df_time['Sched Arrive DateTime'] <= '00:30:00', 'LATE? (Yes/No)'] = 'No' 
    
    return(df_time)

def top10(df, x, y):
    #Top Order
    df = df.nlargest(100000, x)

    #drop duplicate
    df = df.drop_duplicates([y])

    #Top 10
    df = df.nlargest(10, x)
    
    return(df)

time_csv()
```

```python
df_cost = pd.read_csv (r'/Users/josueherrera/Desktop/Towson University/Fall 2021/Data Anayltics Competition/Perdue Farms/data/csv_folder/cost.csv')

#convert Direct Load Cost to float

def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)


def clean_laters(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('S000', '').replace(',', ''))
    return(x)

df_cost['Direct Load Cost'] = df_cost['Direct Load Cost'].apply(clean_currency).astype('float')
df_cost['Pounds Shipped'] = df_cost['Pounds Shipped'].apply(clean_currency).astype('int')
df_cost['Shipment Number'] = df_cost['Gen3,Shipment Number'].apply(clean_laters).astype('int')

#Add $ format
df_cost['Direct Load Cost'] = df_cost['Direct Load Cost']

df_cost
```

```python
#Q1: Which customers would offer the greatest savings by using a drop trailer?

    #First merge TMS & Time df with (Load == Shipment Number)
df_greatsav = pd.merge(time_csv(), tms(), how='inner', left_on=['Sold To', 'Carrier Name', 'Load #'], right_on = ['Dropoff ID', 'Carrier Name', 'Shipment Number'])

df_greatsav
```

```python
#Filter for Perdue & #Stops & Late
df_filter = df_greatsav[df_greatsav['Carrier Name']== 'Perdue']
df_filter = df_filter.loc[df_filter['# of Stops']== 1]

df_filter = df_filter[df_filter['HELD'] >= '01:00:00']

df_filter['# Appearances'] = 1
df_filter = df_filter.nlargest(3000, '# Appearances')

df_filter['HELD (Seconds)'] = df_filter['HELD'] / np.timedelta64(1, 's')

df_filter.loc[df_filter['HELD (Seconds)'] == 0, 'Cost of Appearances'] = 0 
df_filter.loc[df_filter['HELD (Seconds)'] > 0, 'Cost of Appearances'] = df_filter['HELD (Seconds)'] * (65 /3600)



df_filter['# Appearances'] = df_filter.groupby(['Sold To'])['# Appearances'].transform('sum')
df_filter['Total Held Time'] = df_filter.groupby(['Sold To'])['HELD'].transform('sum')

df_filter['HELD (Seconds)'] = df_filter.groupby(['Sold To'])['HELD (Seconds)'].transform('sum')

df_filter['HELD'] = df_filter.groupby(['Sold To'])['HELD'].transform('sum')
df_filter['Cost of Appearances'] = df_filter.groupby(['Sold To'])['Cost of Appearances'].transform('sum')

#Top Order
    #NOTE: top 1500 is the same customers
df_filter = df_filter.nlargest(10000, '# Appearances') 

df_filter = df_filter.drop(df_filter.iloc[:, 8:16], axis=1)
df_filter = df_filter.drop(['Pickup Depart DateTime', 'Sched Arrive DateTime', 'Actual Arrive DateTime', 
                            'Empty DateTime', 'LATE? (Yes/No)', 'Total Held Time', 'Load #'], axis=1)

#df_filter = df_filter.nlargest(1000, '# Appearances')
df_filter = df_filter.drop_duplicates(['Sold To'])

#df_filter = df_filter.nlargest(10, '# Appearances')

df_filter = top10(df_filter, '# Appearances', 'Sold To')

#1500 

df_filter['Cost of Appearances'] = df_filter['Cost of Appearances'].map('${:,.2f} '.format)

df_filter

#Rank by most appearances that are over 1 hr held time. 
#Then ordered by how many time it occurs
#Followed by the total sum of held time per customer
#Divide the entire HELD-Time column by 1 second then multiply by (65 hr/3600 secs(1hr))
#The first 500 shows 
```
